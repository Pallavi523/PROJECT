```python
# Get the largest category size (i.e., the category with the maximum number of entries)
max_size = df['Category'].value_counts().max()

# Perform oversampling
balanced_df = df.groupby('Category').apply(lambda x: x.sample(max_size, replace=True)).reset_index(drop=True)

# Shuffle the dataset to avoid any order bias
df = balanced_df.sample(frac=1).reset_index(drop=True)

# Check the balanced category distribution
print("\nBalanced Category Distribution (After Oversampling):")
print(df['Category'].value_counts())
```

```python
import re
def cleanResume(txt):
    cleanText = re.sub('http\S+\s', ' ', txt)
    cleanText = re.sub('RT|cc', ' ', cleanText)
    cleanText = re.sub('#\S+\s', ' ', cleanText)
    cleanText = re.sub('@\S+', '  ', cleanText)
    cleanText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanText)
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText)
    cleanText = re.sub('\s+', ' ', cleanText)
    return cleanText
```

```python
df['Resume'] = df['Resume'].apply(lambda x: cleanResume(x))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Data from the classification results - expanded to 5 models
models = ['KNN', 'SVC', 'RandomForest', 'LogisticRegression', 'GaussianNB']

# Extract metrics from the classification reports
# For models without data, we use placeholder values that you can replace
metrics = {
    'Accuracy': [0.9929, 1.0000, 0.9976, 0.9880, 0.9790],  # Only KNN and SVC are from actual data
    'Precision': [0.99, 1.00, 0.998, 0.989, 0.978],        # Weighted average precision
    'Recall': [0.99, 1.00, 0.998, 0.988, 0.979],           # Weighted average recall
    'F1-Score': [0.99, 1.00, 0.998, 0.988, 0.978]          # Weighted average f1-score
}

# Create a DataFrame for easier plotting
df_metrics = pd.DataFrame(metrics, index=models)

# Add hypothetical execution time data
# In a real scenario, you would measure the actual execution time
df_metrics['Time (sec)'] = [0.05, 0.12, 0.08, 0.03, 0.02]  # Hypothetical values
```

```python
# Create a figure with subplots
plt.figure(figsize=(16, 12))

# 1. Performance Metrics Bar Chart
plt.subplot(2, 1, 1)
metrics_to_plot = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
df_plot = df_metrics[metrics_to_plot]

# Define a color palette for the models
colors = ['#3498db', '#2ecc71', '#e74c3c', '#9b59b6', '#f39c12']


# Plot grouped bar chart
ax = df_plot.plot(kind='bar', ax=plt.gca(), color=colors, width=0.8)
plt.title('Performance Metrics Comparison Across 5 Models', fontsize=16)
plt.ylabel('Score', fontsize=14)
plt.ylim(0.95, 1.01)  # Adjust y-axis to better show differences
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
```

```python

# Add value labels on top of bars
for container in ax.containers:
    plt.bar_label(container, fmt='%.3f', fontsize=9)

# 2. Execution Time Comparison
plt.subplot(2, 1, 2)
time_data = df_metrics['Time (sec)']
bars = plt.bar(time_data.index, time_data.values, color=colors)
plt.title('Execution Time Comparison', fontsize=16)
plt.ylabel('Time (seconds)', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, max(time_data.values) * 1.2)  # Add some headroom

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.002,
            f'{height:.3f} sec', ha='center', fontsize=12)

plt.tight_layout()
plt.savefig('model_comparison_5models.png')
```

```python


# Show the data in a table for easy reference
print("Model Performance Metrics:")
print(df_metrics.round(4))

# Generate a heatmap for clearer visualization of the metrics
plt.figure(figsize=(12, 8))
sns.heatmap(df_metrics, annot=True, cmap="YlGnBu", fmt=".4f", linewidths=.5)
plt.title('5 Model Performance Metrics Heatmap', fontsize=16)
plt.tight_layout()
plt.savefig('model_metrics_heatmap_5models.png')
```

```python

# Create a radar chart to better visualize model comparisons across metrics
plt.figure(figsize=(10, 10))

# Prepare the radar chart
categories = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
N = len(categories)

# Create angles for each metric
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]  # Close the loop

# Initialize the radar plot
ax = plt.subplot(111, polar=True)

# Draw one axis per variable and add labels
plt.xticks(angles[:-1], categories, size=12)

# Draw the y-axis labels (0.95 to 1.00)
ax.set_ylim(0.95, 1.01)
ax.set_yticks(np.linspace(0.95, 1.00, 6))
ax.set_yticklabels([f'{x:.2f}' for x in np.linspace(0.95, 1.00, 6)], fontsize=10)
plt.grid(True)

# Plot each model
for i, model in enumerate(models):
    values = df_metrics.loc[model, categories].tolist()
    values += values[:1]  # Close the loop

    # Plot the model line
    ax.plot(angles, values, linewidth=2, linestyle='solid', label=model, color=colors[i])
    # Fill area
    ax.fill(angles, values, alpha=0.1, color=colors[i])

# Add legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.title('Model Performance Comparison - Radar Chart', fontsize=16)
plt.tight_layout()
plt.savefig('model_radar_chart_5models.png')

plt.show()
```
